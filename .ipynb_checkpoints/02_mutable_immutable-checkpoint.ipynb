```python
#Mutable objects: whose value can be changed after they are created.
# eg list, dictiionary 
list_01 = [ "ram" , "hari" , 54 , 45.3, True ]
list_01.append(45)
print(list_01) 
```

```python
my_list = [1, 2, 2, 3]
print(my_list.count(2))
print(my_list.index(2))
```

```python
#Immutable objects: whose value can't be changed after they are created.
# for eg: str , tuples
my_str = "hari"
my_str.append("shyam")
print(my_str) # it will provide error 
```
